## Migrando dados da Planilha para um script de inser√ß√£o

#### Preparando ambiente

```python
import pandas as pd
import random
```

```python
caminho_arquivo_escrita = '../data/gerados/dml.sql'
```

```python
df_planilha = pd.read_excel("../data/eventify.xlsx", header=0)
df_planilha.head()
```

#### Eliminando o Mural

```python
qtd_antes = df_planilha.shape[0]
df_planilha = df_planilha[df_planilha["Nome do Bucket"] != "üìå Mural"]
qtd_depois = df_planilha.shape[0]
print(f"Antes: {qtd_antes} | Depois: {qtd_depois}")
```

#### Eliminando colunas indesejadas

```python
df_planilha.drop(columns=["Identifica√ß√£o da tarefa", "Prioridade", "√â Recorrente", "Itens conclu√≠dos da lista de verifica√ß√£o", "Itens da lista de verifica√ß√£o"], inplace=True)
df_planilha.head()
```

#### Limpando os caracteres especiais do campo Descri√ß√£o

```python
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("\n", " ")
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("\r", " ")
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("\t", " ")
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("_x000d_", "")

df_planilha.head()
```

#### Limpando poss√≠veis quebras de Script

```python
df_planilha = df_planilha.astype(str).applymap(lambda x: x.replace("'", "").replace('"', ""))
df_planilha.head()
```

#### Limitando descri√ß√µes muito longas

```python
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.slice(0, 256)
df_planilha.head()
```

#### Limpando R√≥tulos para se tornar Fibonacci

```python
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("[^0-9]", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("¬™", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("¬∫", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace(";5", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace(";AguardarProfessor", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("Aguardar Professor", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace(";", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("Sprint", "")
df_planilha.head()
```

```python
df_planilha["R√≥tulos"] = pd.to_numeric(df_planilha["R√≥tulos"], errors="coerce").fillna(0).astype(int)
df_planilha.head()
```

#### Criando o Board e o Section

```python
 boards_sections = df_planilha['Nome do Bucket'].unique()

df_planilha = df_planilha.assign(board="", section="")
df_planilha.head()

for index, row in df_planilha.iterrows():
    if "Back-end" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Back-end"
    
    if "Rest-API" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Back-end"
    
    if "Back-End" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Back-end"

    if "C√≥digo" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Front-end"
    
    if "Front-end" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Front-end"
    
    if "Web-App" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Front-end"

    if "Front-End" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Front-end"
    
    if "Te√≥rico" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Te√≥rico e Outros"
    
    if "Infra" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Te√≥rico e Outros"

    if "Outros" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Te√≥rico e Outros"
    
    if "Checkout" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Te√≥rico e Outros"

    if "Documenta√ß√£o" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Te√≥rico e Outros"

    if "SPRINT IV" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT IV" 
    elif "SPRINT III" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT III"
    elif "SPRINT II" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT II"
    elif "SPRINT I" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT I"
    elif "Refatora√ß√£o" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "Refatora√ß√£o"

df_planilha.head()
```

#### Criando a coluna status a partir do campo "Conclu√≠da em"

```python
df_planilha["status"] = df_planilha["Conclu√≠do em"].apply(lambda x: 2 if pd.notnull(x) else 0)
df_planilha.head()
```

#### Criando a coluna time a partir de "R√≥tulos"

```python
df_planilha["time"] = pd.to_numeric(df_planilha["R√≥tulos"], errors='coerce').fillna(0)
df_planilha["time"] = df_planilha["time"] / 2
df_planilha["time"] = df_planilha["time"].astype(int)

df_planilha.head()
```

#### Criando a coluna Priority a partir de "R√≥tulos"

```python
# Se R√≥tulos for menor igual a 3, prioridade 0
# Se R√≥tulos for maior que 3 e menor que 13, prioridade 1
# Se R√≥tulos for maior igual que 13, prioridade 2
# Se R√≥tulos for vazio, prioridade 0

df_planilha["prioridade"] = pd.to_numeric(df_planilha["R√≥tulos"], errors='coerce').fillna(0)
df_planilha["prioridade"] = df_planilha["prioridade"].apply(lambda x: 0 if x <= 3 else 1 if x < 13 else 2)

df_planilha.head()
```

#### Criando a coluna data_estimada a partir de time

```python
df_planilha["Criado em"] = pd.to_datetime(df_planilha["Criado em"], format="%d/%m/%Y")
df_planilha["data_estimada"] = df_planilha["Criado em"] + pd.to_timedelta(df_planilha["time"], unit='d')
df_planilha["data_estimada"] = df_planilha["data_estimada"].dt.strftime("%Y-%m-%d")
df_planilha.head()
```

#### Simulando a Coluna Data Conclus√£o

```python
# Transforme a coluna "data_estimada" em datetime
df_planilha["data_estimada"] = pd.to_datetime(df_planilha["data_estimada"], format="%Y-%m-%d")
df_planilha.head()
```

```python
# Se a Conclu√≠do em for nula, Coclu√≠do em √© igual a data estimada - valor randomico entre 3 e 10 dias

df_planilha["Conclu√≠do em"] = pd.to_datetime(df_planilha["Conclu√≠do em"], format="%d/%m/%Y")
df_planilha["Conclu√≠do em"] = df_planilha["Conclu√≠do em"].fillna(df_planilha["data_estimada"] - pd.to_timedelta(random.randint(3, 10), unit='d'))
df_planilha["Conclu√≠do em"] = df_planilha["Conclu√≠do em"].dt.strftime("%Y-%m-%d")
df_planilha.head()
```

#### Criando uma coluna de ID para montar os relacionamentos

```python
df_planilha = df_planilha.assign(id=range(1, len(df_planilha) + 1))
df_planilha.head()
```

#### Atribuindo a cada usu√°rio a tarefa

- [1] BRAYAN COELHO DE AQUINO
- [2] GUILHERME HENRIQUE ANGELO DIAS
- [3] HELDER DAVIDSON RODRIGUES ALVARENGA
- [4] LEONARDO VASCONCELOS PAULINO

```python
df_usuario_tarefa = pd.DataFrame(columns=['id_usuario', 'id_tarefa'])

for index, row in df_planilha.iterrows():
    atribuido_a = str(row["Atribu√≠do a"])
    if "BRAYAN COELHO DE AQUINO" in atribuido_a:
        df_usuario_tarefa = pd.concat([df_usuario_tarefa, pd.DataFrame({"id_usuario": 1, "id_tarefa": row["id"]}, index=[0])], ignore_index=True)
    
    if "GUILHERME HENRIQUE ANGELO DIAS" in atribuido_a:
        df_usuario_tarefa = pd.concat([df_usuario_tarefa, pd.DataFrame({"id_usuario": 2, "id_tarefa": row["id"]}, index=[0])], ignore_index=True)
    
    if "HELDER DAVIDSON RODRIGUES ALVARENGA" in atribuido_a:
        df_usuario_tarefa = pd.concat([df_usuario_tarefa, pd.DataFrame({"id_usuario": 3, "id_tarefa": row["id"]}, index=[0])], ignore_index=True)
    
    if "LEONARDO VASCONCELOS PAULINO" in atribuido_a:
        df_usuario_tarefa = pd.concat([df_usuario_tarefa, pd.DataFrame({"id_usuario": 4, "id_tarefa": row["id"]}, index=[0])], ignore_index=True)
    
    if "VICTOR HUGO MARQUES DO NASCIMENTO" in atribuido_a:
        df_usuario_tarefa = pd.concat([df_usuario_tarefa, pd.DataFrame({"id_usuario": 5, "id_tarefa": row["id"]}, index=[0])], ignore_index=True)
    
    if "VITOR GON√áALVES DA SILVA" in atribuido_a:
        df_usuario_tarefa = pd.concat([df_usuario_tarefa, pd.DataFrame({"id_usuario": 6, "id_tarefa": row["id"]}, index=[0])], ignore_index=True)

df_usuario_tarefa.head()
```

#### Impedindo que entre um Fibonacci 0 ou nulo

```python
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].apply(lambda x: 5 if x == 0 else x)
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].fillna(5)
df_planilha.head()
```

#### Criando a tabela de Board

```python
df_boards = pd.DataFrame(columns=['id', 'name'])
df_boards = df_planilha[['board']].drop_duplicates().reset_index(drop=True)
df_boards = df_boards.rename(columns={"board": "name"})
df_boards = df_boards[df_boards["name"] != ""]
df_boards = df_boards.assign(id=range(1, len(df_boards) + 1))
df_boards.head()
```

#### Criando a tabela de Sections

```python
data = ["SPRINT I", 1], ["SPRINT II", 2], ["SPRINT III", 3], ["Refatora√ß√£o", 4], ["SPRINT IV", 5]
df_sections_ordenados = pd.DataFrame(data, columns=['name', 'id'])
df_sections_ordenados.head()
```

#### Criando a tabela de sections relacionada ao board

```python
df_sections = pd.DataFrame(columns=['id', 'name', 'board_id'])

for index, row in df_sections_ordenados.iterrows():
    for index2, row2 in df_boards.iterrows():
        df_sections = pd.concat([df_sections, pd.DataFrame({"id": row["id"], "name": row["name"], "board_id": row2["id"]}, index=[0])], ignore_index=True)

df_sections["id"] = df_sections.index + 1
df_sections.head(20)
```

#### Criando a tabela de tarefas a partir da planilha

```python
df_tarefas = pd.DataFrame(columns=['id', 'name', 'description', 'priority', 'fibonacci', 'status', 'time', 'data_estimada', 'section_id', 'date_creation'])

for index, row in df_planilha.iterrows():
    df_tarefas = pd.concat([df_tarefas, pd.DataFrame({"id": row["id"], "name": row["Nome da tarefa"], "description": row["Descri√ß√£o"], "priority": row["status"], "fibonacci": row["R√≥tulos"], "status": row["status"], "time": row["time"], "data_estimada": row["data_estimada"], "section_id": row["section_id"], "date_creation": row["Criado em"]}, index=[0])], ignore_index=True)

df_tarefas.head()
```

#### Cria√ß√£o de Script SQL

```python
txt = "INSERT INTO eventitask.board (name) VALUES "

for index, row in df_boards.iterrows():
    txt += f"\n    ('{row['name']}'),"

txt = txt[:-1] + ";"
print(txt)

with open(caminho_arquivo_escrita, "a", encoding='utf-8') as arquivo:
    arquivo.write(txt)
```

```python
txt = "\n\nINSERT INTO eventitask.section (name, board_id) VALUES "

for index, row in df_sections.iterrows():
    txt += f"\n    ('{row['name']}', {row['board_id']}),"

txt = txt[:-1] + ";"
print(txt)

with open(caminho_arquivo_escrita, "a", encoding='utf-8') as arquivo:
    arquivo.write(txt)
```

```python
txt = "\n\nINSERT INTO eventitask.task (section_id, name, description, priority, fibonacci, status, time, data_estimada, date_creation) VALUES "

for index, row in df_tarefas.iterrows():
    txt += f"\n    ({row['section_id']}, '{row['name']}', '{row['description']}', {row['priority']}, '{row['fibonacci']}', {row['status']}, {row['time']}, '{row['data_estimada']}', '{row['date_creation']}'),"

txt = txt[:-1] + ";"
print(txt)

with open(caminho_arquivo_escrita, "a", encoding='utf-8') as arquivo:
    arquivo.write(txt)
```

```python
txt = "\n\nINSERT INTO eventitask.user_task (user_id, task_id) VALUES "

for index, row in df_usuario_tarefa.iterrows():
    txt += f"\n    ({row['id_usuario']}, {row['id_tarefa']}),"

txt = txt[:-1] + ";"
print(txt)

with open(caminho_arquivo_escrita, "a", encoding='utf-8') as arquivo:
    arquivo.write(txt)
```

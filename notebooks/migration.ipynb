## Migrando dados da Planilha para um script de inser√ß√£o

#### Preparando ambiente

```python
import pandas as pd
```

```python
df_planilha = pd.read_excel("../data/eventify.xlsx", header=0)
df_planilha.head()
```

#### Eliminando o Mural

```python
qtd_antes = df_planilha.shape[0]
df_planilha = df_planilha[df_planilha["Nome do Bucket"] != "üìå Mural"]
qtd_depois = df_planilha.shape[0]
print(f"Antes: {qtd_antes} | Depois: {qtd_depois}")
```

#### Limpando os caracteres especiais do campo Descri√ß√£o

```python
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("\n", " ")
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("\r", " ")
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("\t", " ")
df_planilha["Descri√ß√£o"] = df_planilha["Descri√ß√£o"].str.replace("_x000d_", "")

df_planilha.head()
```

#### Limpando R√≥tulos para se tornar Fibonacci

```python
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("[^0-9]", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("¬™", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("¬∫", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace(";5", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace(";AguardarProfessor", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("Aguardar Professor", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace(";", "")
df_planilha["R√≥tulos"] = df_planilha["R√≥tulos"].str.replace("Sprint", "")
df_planilha.head()
```

#### Criando o Board e o Section

```python
boards_sections = df_planilha['Nome do Bucket'].unique()

df_planilha = df_planilha.assign(board="", section="")
df_planilha.head()

for index, row in df_planilha.iterrows():
    if "Back-end" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Back-end"
    elif "Rest-API" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Back-end"
    elif "Front-end" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Front-end"
    elif "Web-App" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Front-end"
    elif "Te√≥rico" in row["Nome do Bucket"]:
        df_planilha.at[index, "board"] = "Te√≥rico e Outras"
    else:
        df_planilha.at[index, "section"] = "Te√≥rico e Outras"
    
    if "SPRINT IV" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT IV"
    elif "SPRINT III" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT III"
    elif "SPRINT II" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT II"
    elif "SPRINT I" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "SPRINT I"
    elif "Refatora√ß√£o" in row["Nome do Bucket"]:
        df_planilha.at[index, "section"] = "Refatora√ß√£o"

df_planilha.head()
```

#### Criando a coluna status a partir do campo "Conclu√≠da em"

```python
df_planilha["status"] = df_planilha["Conclu√≠do em"].apply(lambda x: 2 if pd.notnull(x) else 0)
df_planilha.head()
```

#### Criando a coluna time a partir de "R√≥tulos"

```python
df_planilha["time"] = df_planilha["R√≥tulos"].apply(lambda x: int(x) if pd.notnull(x) else 0)
df_planilha["time"] = df_planilha["time"].apply(lambda x: x/2)
df_planilha["time"] = df_planilha["time"].apply(lambda x: int(x) if pd.notnull(x) else 0)
df_planilha.head()
```

#### Criando a coluna data_estimada a partir de time

```python
df_planilha["Criado em"] = pd.to_datetime(df_planilha["Criado em"], format="%d/%m/%Y")
df_planilha["data_estimada"] = df_planilha["Criado em"] + pd.to_timedelta(df_planilha["time"], unit='d')
df_planilha["data_estimada"] = df_planilha["data_estimada"].dt.strftime("%Y-%m-%d")
df_planilha.head()
```

#### Criando uma coluna de ID para montar os relacionamentos

```python
df_planilha = df_planilha.assign(id=range(1, len(df_planilha) + 1))
df_planilha.head()
```
